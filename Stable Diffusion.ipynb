## **Downloading and importing the needed libraries**

```python
#!pip install --upgrade diffusers accelerate transformers
from diffusers import DiffusionPipeline
```

## **Preparing the Pre-trained model**

```python
pipeline = DiffusionPipeline.from_pretrained("digiplay/Realisian_v5")
pipeline.to("cuda")
```

## **Running the model**

```python
prompt = "alien riding a white horse in jupiter in a tornado"
image = pipeline(prompt).images[0]
image
```

Saving the image

```python
image.save(f"{prompt}.png")
```
